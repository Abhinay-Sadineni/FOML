Develop an ordinal regression model in Python to predict the rating of the wine based on the input features.You may consider 20% of the data for testing purpose.Also, propose an appropriate evaluation metric and compare the result with a linear regression model.You may use existing ordinal regression and linear regression packages 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from mord import LogisticIT
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import  mean_squared_error, mean_absolute_error , r2_score  , accuracy_score
```

<center><h2>White wine quality prediction</h2></center>

```python
# Create a dataframe with all columns
# Create X as all columns except quality  
# Create y as quality
df = pd.read_csv("wine+quality/winequality-white.csv", delimiter=";")
X = df.drop("quality", axis=1)
y = df["quality"]
```

```python
# create training and testing , with training 80% and testing 20%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
#display training data
X_train.head()
```

```python
#display testing data 
X_test.head()
```

```python
#Train linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)


#Train LogisticIT model(ordinal regression)
ordinal_model = LogisticIT()
ordinal_model.fit((X_train),y_train)
#Train LogisticRegression model


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

logistic_model = LogisticRegression(max_iter=300)
logistic_model.fit((X_train_scaled), y_train)
```

```python
# Test the models
linear_predictions = linear_model.predict(X_test)
ordinal_predictions = ordinal_model.predict(X_test.values)
logistic_predictions = logistic_model.predict(X_test_scaled)
```

```python
# Compare the models
linear_MSE = mean_squared_error(y_test, linear_predictions)
ordinal_MSE = mean_squared_error(y_test, ordinal_predictions)

linear_MAE = mean_absolute_error(y_test, linear_predictions)
ordinal_MAE = mean_absolute_error(y_test, ordinal_predictions)

ordinal_accuracy = accuracy_score(y_test, ordinal_predictions)
logistic_accuracy = accuracy_score(y_test, logistic_predictions)
```

```python
print(f"Linear Regression evaluation metric(MSE): {linear_MSE}")
print(f"ordinal Regression evaluation metric(MSE): {ordinal_MSE}")

print(f"Linear Regression evaluation metric(MAE): {linear_MAE}")
print(f"ordinal Regression evaluation metric(MAE): {ordinal_MAE}\n")

print(f"ordinal Regression evailuation metric(accuracy): {ordinal_accuracy}")
print(f"logistic Regression evailuation metric(accuracy): {logistic_accuracy}")
```
